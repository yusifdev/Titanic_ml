```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")
print("Path to dataset files:", path)
```

```python
import pandas as pd
import os
df = pd.read_csv(os.path.join(path, "Titanic-Dataset.csv"))
display(df) # or df.head(len(df))
#This displays the whole dataset of Titanic on kagglehub.
```

```python
display(df['Survived'])
#This just shows the survived column.
```

```python
survived_people=0
for person_number in df['Survived']:
  if person_number==1:
    survived_people+=1
print(survived_people)
#This shows the number of survived people on Titanic event.(Task 1, first part)
```

```python
print(df['Fare'].describe()) #statistical important features of 'fare' column
mean=df['Fare'].mean()
print(mean) #mean of 'fare' column
median=df['Fare'].median() #median2=df['Fare'].quantile(0.5)
#alternative median function
print(median) #median of 'fare' column
#This shows the mean and median of ticket prices on Titanic event.
#Task 1, second part
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array(df['SibSp'])
ypoints=np.array(df['Parch'])
plt.scatter(xpoints,ypoints)
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Number of Parents/Children Aboard')
plt.title('Relationship between the numbers of siblings/spouses and parents/children abroad')
plt.show()
#This plots the relationship between the numbers of siblings/spouses and parents/children abroad.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#both libraries are used for making histograms
# Using matplotlib.pyplot.hist()
plt.figure(figsize=(10, 5))#chart sizes(width*height)
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black')#dropna=drop NaN values, bins=20 means 20 equal bars, edgecolor=black means the edgecolor of bars are black
plt.title('Distribution of Age (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Frequency')#chart and axis title names
plt.grid(axis='y', alpha=0.75)#grid system based on which axis(y means from y axis to left or right, x means from x axis to top or bottom) and alpha means their opacity.
plt.show()

# Using seaborn.histplot() for a more aesthetically pleasing plot
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True, edgecolor='black')#kde True means it will show the curve line, false means no curve line will be shown.
plt.title('Distribution of Age (Seaborn)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
#This plot shows the frequency of people from different ages by using two methods.
#Task 1, third(bonus) part
```

This code generates two histograms for the 'Age' column:

*   The first one uses `plt.hist()` from `matplotlib`, which is a basic way to create a histogram.
*   The second one uses `sns.histplot()` from `seaborn`, which is built on top of Matplotlib and often provides more visually appealing plots and additional features like the Kernel Density Estimate (`kde=True`).

I've used `.dropna()` to exclude any missing age values from the plot, and `bins=20` to define the number of bars in the histogram.
